```python
k = 0.3
totpers=257805
npp_viv=totpers/len(data0)

for ed in data0["edo"].unique():
    countpe = 0
    countpobre = 0
    sumape = 0
    sumapobre = 0
    vivedo=0
    for a in range(len(data0)):
        if(data0["edo"][a]==ed):
            vivedo = vivedo+1
            if(data0["c"][a] >0):
                countpobre = countpobre + 1
                sumapobre = sumapobre + data0["c"][a]
            if (data0["c"][a] >= k ) :
                countpe = countpe + 1
                sumape =sumape + data0["c"][a]
    Hp= countpobre/vivedo
    Ap = sumapobre/countpobre
    MEPIp= Ap*Hp
    Hpe = countpe/vivedo
    Ape = sumape/countpe
    MEPIpe=Hpe*Ape
    print("Resultados estado:", ed, "H y MEPI", Hp,MEPIp, "Hpe y MEPI pe", Hpe, MEPIpe)
    print("viviendas por estado", vivedo)
    mepi2016.append(MEPIp)
    vivpob2016.append(Hp)
    vivpobex2016.append(Hpe)
    estados.append(ed)
```
